## DropViz query
**Limit By Region:** Striatum  
**Limit By Class:** Oligodendrocytes, Neurons

| Label | Name | Marker |
| :---  | :--- | :---   |
| OL    | oligodendrocyte | Tfr |
| dSPN_Drd1 | direct pathway spiny projection neuron | Drd1 |
| iSPN_Adora2a | indirect pathway spiny projection neuron | Adora2a |
| Int_Chat | cholinergic interneuron | ChAT |
| Neu_SPN | spiny projection neurons of <br>bed nucleus of the stria <br>terminalis (BNST), amygdala | Otof |
| Int_Pvalb | parvalbumin interneuron | Pnoc |
| Int_Sst | somatostatin interneuron | (SST) |

```python
import numpy as py
import pandas as pd

rawdata = pd.read_csv("DropViz_DEGs_TP100k.csv")

data = rawdata.set_axis(['gene','log2FC','OL','dSPN_Drd1','iSPN_Adora2a','Int_Chat','Neu_SPN','Int_Pvalb','Int_Sst'], axis=1, inplace=False)
data['FC_mag'] = data['log2FC'].abs()

ranked = data.sort_values(['OL','FC_mag'], ascending=[False,False])
ranked = ranked[['gene','log2FC','OL','dSPN_Drd1','iSPN_Adora2a','Int_Chat','Neu_SPN','Int_Pvalb','Int_Sst']]
ranked.reset_index(drop=True, inplace=True)
ranked
```
